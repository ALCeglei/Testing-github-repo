```python
import matplotlib.pyplot as plt
from matplotlib import style
import numpy as np

style.use('ggplot')

x = np.arange(-20, 20, 0.1)
y = x**2
x2 = np.arange(-20, 20, 0.1)
y2 = x*10 - x**2
plt.subplot(221)
plt.title('See differences')
plt.plot(x, y,'g',label='quadratic func.', linewidth=4)
plt.xlabel('X Axis')
plt.ylabel('Y Axis')
plt.legend()
plt.grid(True, color='k')

plt.subplot(222)

a = np.arange(-20, 20, 0.1)
b = x**2 + x/10
plt.plot(a,b,'b',label='x**2 + x/10',linewidth=3)
plt.xlabel('X Axis')
plt.ylabel('Y Axis')
plt.legend()

plt.subplot(223)  # 2 rows, 2 columns, this is third

plt.plot(x2,y2,'b',label='x*10 - x**2',linewidth=3)
plt.xlabel('X Axis')
plt.ylabel('Y Axis')
plt.legend()
plt.grid(True, color='k')
plt.show()
```

```python
age = np.array([17, 26, 38, 20, 48])
salary = np.array([600, 2300, 5000, 1750, 8000])
age2 = np.array([19, 23, 30, 25, 50])
salary2 = np.array([1000, 2000, 3170, 2300, 7000])



plt.bar(age,salary, label='men\'s salary', color='b', width=1.5)
plt.bar(age2, salary2, label='women\'s salary', color='r', width=1.5)
plt.legend()
plt.title('salary')
plt.xlabel('age')
plt.ylabel('dollars')
plt.show()
```

```python
animals = np.array(['dog', 'cat', 'mouse', 'rabbit', 'hawk', 'parrot','parrot', 'cat', 'cat', 'dog', 'mouse'])
bins = np.array([1, 2, 3, 4, 5, 6, 7])

plt.hist(animals, bins,label='animals', histtype='bar', width=0.55, color='b')
plt.ylabel('Quontity of animals')
plt.xlabel('Animals')
plt.title('Count of home animals')
plt.legend()
plt.show()
```

```python
x = [1,2,9,5,3,6,7,8,3,5,6,7,5,3]
y = [9,4,8,3,2,4,3,2,1,4,6,8,3,2]

plt.scatter(x,y,label='just points', color='green')
plt.title('Scatter plot')
plt.xlabel('X')
plt.ylabel('Y')
plt.legend()
plt.show()
```

```python
days = [1, 2, 3, 4, 5]
play = [3, 5, 6, 2, 4]
eat = [1, 2, 4, 1.2, 1]
sleep = [6, 7, 5, 8, 7]

plt.plot([], [], color='m', label='play', linewidth=5)
plt.plot([], [], color='g', label='eat', linewidth=5)
plt.plot([], [], color='b', label='sleep', linewidth=5)

plt.legend()

plt.title('Strange... area plot')

plt.stackplot(days, play, eat, sleep, colors=['m', 'g', 'b'])

plt.show()
```

```python
slices = [3, 8, 4, 6, 3]
cols = ['g', 'b', 'm', 'r', 'y']
activities = ['eat', 'sleep', 'study', 'work', 'rest']

plt.title('Pie Plot')
plt.pie(slices, labels = activities,
       colors = cols, startangle=90,
       shadow=True, explode=(0.05,0.09,0.05,0.05,0.05),
       autopct='%1.2f%%')
```
